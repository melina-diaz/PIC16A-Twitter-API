Column Descriptions:
- text: Text of the tweet
- id_x: unique numerical ID of the tweet
- created_at_x: datetime of when tweet was created
- referenced tweets: info if the tweet is a retweet or reply to another tweet
- geo: id of location where the tweet was created
- retweet_count_x: number of retweets of this tweet
- reply_count_x: number of replies to this tweet
- like_count_x: number of likes of this tweet
- description: description of author account
- id_y: unique numerical ID of author
- verified: bool if the author is verified
- username: unique username, preceded by @
- created_at_y: datetime when the author account was created
- name: name of author
- followers_count_y: number of users following author
- following_count_y: number of users author is following
- tweet_count_y: total number of tweets author has posted
- listed_count_y: the number of public lists users have author to
